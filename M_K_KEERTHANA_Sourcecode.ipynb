```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/Netflix Dataset (10).csv')
```

```python
# Basic analysis and structure understanding
print("Shape of dataset:", df.shape)            # rows & columns
print("\nColumn names:", df.columns.tolist())   # list all columns

print("\nDataset Info:")
df.info()                                       # data types & nulls

print("\nMissing values in each column:")
print(df.isnull().sum())                        # count missing values

print("\nUnique categories in 'Category' (Movie vs TV Show):")
print(df['Category'].unique())

print("\nUnique Ratings (like PG, TV-MA, etc):")
print(df['Rating'].unique()[:10])               # first 10 unique values

print("\nFirst 5 rows of dataset:")
df.head()
```

```python
print("Shape of dataset:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nDataset Info:")
df.info()
print("\nMissing values in each column:")
print(df.isnull().sum())
print("\nUnique categories in 'Category':")
print(df['Category'].unique())
print("\nUnique Ratings:")
print(df['Rating'].unique()[:10])
print("\nFirst 5 rows of dataset:")
df.head()
```

```python
# Convert 'Release_Date' to proper datetime format
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')

# Drop duplicate rows if any
df = df.drop_duplicates()

# Fill missing values in Director, Cast, and Country with "Not Specified"
df['Director'] = df['Director'].fillna("Not Specified")
df['Cast'] = df['Cast'].fillna("Not Specified")
df['Country'] = df['Country'].fillna("Not Specified")

# Confirm the cleaning
print("Missing values after cleaning:")
print(df.isnull().sum())
```

1) Movies vs TV Shows Count

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Category', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

2) Content Added Over the Years

```python
df['Year'] = df['Release_Date'].dt.year

df['Year'].value_counts().sort_index().plot(kind='bar', figsize=(12,5))
plt.title("Content Added per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

3) Top 10 Countries with Most Content

```python
df['Country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

4) Most Common Genres / Types

```python
df['Type'] = df['Type'].astype(str)
genres = df['Type'].str.split(', ').explode()
genres.value_counts().head(10).plot(kind='bar', figsize=(12,5))
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

Ratings Distribution
